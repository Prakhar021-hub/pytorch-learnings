```python
import torch
```

```python
x = torch.tensor(3.0 , requires_grad=True)
```

```python
y = x**2
```

```python
print(y)
```

```python
y.backward()
```

```python
x.grad   ## this is the way to see the gradient (dy/dx)
```

```python
## with chain rule
```

```python
x = torch.tensor(4.0, requires_grad=True)
```

```python
y = x**3
```

```python
z = torch.sin(y)
```

```python
z.backward()
```

```python
x.grad
```
