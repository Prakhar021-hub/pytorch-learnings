## mathematical operations on tensors


```python
## 1. scalar operation

torch.manual_seed(3)
y = torch.rand(2,2)
y
```

```python
## addition
print(y + 3)

## substractions
print(y-1)

## multiplication
print(y*2)

## division
print(y/2)

## power
print(y**2)
```

### performing mathematical operations/element wise operations on two identical tensor

```python
a = torch.rand(2,2)
b = torch.rand(2,2)
print(a)
print(b)
```

```python
## addition
print(a + b )

## substraction
print(a-b)

## multiplication
print(a*b)

## division
print(a/b)
```

```python
## reduction operation

c = torch.randint(size = (3,3), low = 1, high = 5)
c
```

```python
## sum
print(torch.sum(c))

## sum along column
print(torch.sum(c, axis = 0))

## sum along row
print(torch.sum(c, axis = 1))
```

```python
# mean
print(torch.mean(c, dtype = torch.float32))

## mean along colomn
print(torch.mean(c, axis = 0, dtype = torch.float32))

## mean along row
print(torch.mean(c, axis = 1, dtype = torch.float32))
```

```python
## median
print(torch.median(c))
```

```python
## min / max

print(torch.min(c))
print(torch.max(c))
```

```python
## product
torch.prod(c)
```

```python
## standard deviation
torch.std(c.float())
```

```python
## variance
torch.var(c.float())
```

## matrix operations

```python
g = torch.randint(size = (2,3), low = 2, high = 10)
h = torch.randint(size = (2,3), low = 2, high = 10)
print(g)
print(h)
```

```python
## matrix multiplication
print(torch.matmul(g, h.T))
```

```python
## dot product
print(torch.tensordot(g, h.T, dims = 1))
```

```python
## vector dot products
vector1 = torch.tensor([2,4])
vector2 = torch.tensor([3,7])
print(vector1)
print(vector2)

print(torch.dot(vector1, vector2))
```
